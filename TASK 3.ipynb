```python
# Scatter Plot with year against own

plt.scatter(data['year'],data['own'])

# Adding title to the Plot

plt.title("SCATTER PLOT")

# Setting x and y lable

plt.xlabel('YEAR')
plt.ylabel('OWN')

# Showing the result

plt.show()
```

```python
# Line Char against year with own_prop

plt.plot(data['year'],data['own_prop'])

# Add title to the plot

plt.title("LINE CHART")

# Setting x and y labels

plt.xlabel('YEAR')
plt.ylabel('OWN_PROP')

#Showing the result

plt.show()
```

```python
# Bar Chart year against income

plt.bar(data['year'],data['income'])

# Adding Title To the Plot

plt.title("BAR CHART")

# Seeting x and y lable

plt.xlabel('YEAR')
plt.ylabel('INCOME')

# Sowing the result

plt.show()
```

```python
# Histogram

plt.hist(data['income'])

# Adding title to Plot

plt.title('HISTOGRAM')

# Showing the result

plt.show()
```
